```python
import pandas as pd
mengajar = pd.read_csv('/content/mengajar.csv')
dosen = pd.read_csv('/content/tb_dosen.csv')
kelas = pd.read_csv('/content/tb_kelas.csv')
krs = pd.read_csv('/content/tb_krs.csv')
mahasiswa = pd.read_csv('/content/tb_mahasiswa.csv')
matkul = pd.read_csv('/content/tb_mata_kuliah.csv')
nilai = pd.read_csv('/content/tb_nilai.csv')

nilai_dosen = nilai.groupby(nilai['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dosen, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nilai.groupby(nilai['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

Interpretasi : 
- mean_nilai : rata-rata dari keseluruhan nilai yang ada pada data
- std_nilai : simpangan baku
- kita dapat mengetahui kecenderungan dosen dalam memberikan nilai kepada mahasiswa adalah dari tinggi rendahnya mean_nilai
- pada data di atas, mean_nilai terendah didapat oleh bu Annisa Fitria yaitu 60.090909 dan standar deviasi 16.220077. maka dapat diasumsikan bahwa bu Annisa Fitria merupakan dosen yang cenderung memberikan nilai rendah
sedangkan, mean_nilai tertinggi didapat oleh pak Andri Basuki yaitu 85.750000 dan standar deviasi 7.741355 . maka dapat diasumsikan bahwa Pak Andri Basuki merupakan dosen yang cenderung memberikan nilai tinggi
